```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark").getOrCreate()
```

```python
from google.colab import drive
drive.mount("/content/drive/")
```

```python
#reads the content of the file
bank = spark.sparkContext.textFile("drive/My Drive/bank.txt")
```

```python
#it returns a list
bank.collect()
```

```python
bankDetail = bank.flatMap(lambda d:d.split('\t'))
```

```python
bankDetail.collect()
```

```python
bankDetail.map(lambda d:(d,1)).reduceByKey(lambda a,b:a+b).collect()
```

```python
def getTransactionCountForAccount(data):
    if len(data)>=9 and '-' not in data:
      return data
```

```python
account=bankDetail.map(lambda d:getTransactionCountForAccount(d)).filter(lambda d:d!=None)
```

```python
account.map(lambda d: (d, 1)).reduceByKey(lambda a, b: a + b).collect()
```

```python
def getNumberOFAccounts(data):
  if len(data)==8:
    return data
```

```python
account=bankDetail.map(lambda d:getNumberOFAccounts(d)).filter(lambda d:d is not None)
```

```python
#Getting number of transactions per a particular bank
account.map(lambda d:(d,1)).reduceByKey(lambda a,b:a+b).collect()
```

```python
#Total transacted amount per bank
bank.map(lambda d:(d.split('\t')[0],int(d.split('\t')[4]))).reduceByKey(lambda a,b:a+b).collect()
```
