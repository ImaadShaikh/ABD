```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("spark").getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
student = spark.sparkContext.textFile('drive/My Drive/Student.txt')
```

Create dataset (text file) with fields like ‘Student Name’, ‘Institute’, ‘Program Name’, and ‘Gender’ and solve following questions

```python
student.collect()
```

Q2) Compute number of students from each Institute.


```python
student.map(lambda data : (data.split(" ")[1],1)).reduceByKey(lambda a,b: a+b).collect()
```

q3)Number of students enrolled to any program.

```python
student.map(lambda data : (data.split(" ")[2],1)).reduceByKey(lambda a,b :a+b).collect()
```

Q4) Number of ‘boy’ and ‘girl’ students.

```python
student.map(lambda data : (data.split(" ")[3],1)).reduceByKey(lambda a,b : a+b).collect()
```
